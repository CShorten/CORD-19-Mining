```python
# Atomic_Unit_Construction
'''
This notebook explores some ways to break up the large CORD-19 papers 
into smaller units. These smaller units refer to collections that can be 
input to a Transformer neural network. 

Long sequence lengths come back to bite us in the ass computationally, 
so we need to compress this as much as possible.
'''
print("Atomic Unit Construction")
```

```python
# Strategy #1: Cut each text in the body paragraph to length k
```

```python
# First do this with the one paper, then loop through the folder
# Could probably store this as a DataFrame
import json
import pandas as pd
import os

seqs = []
labels = [] # we label them by the paper index for the sake of contrastive sampling
label = 1

k = 128 # cut each body text to length k

DIR = 'pdf_json_1'

for path in os.listdir(DIR):
    with open(DIR+'/'+path) as jsonfile:
        paper_dict = json.load(jsonfile)
    
        body_texts = paper_dict['body_text']
        body_texts_count = len(body_texts)

        for i in range(body_texts_count):
            sample = body_texts[i]['text'].split(' ')
            new_sample = ''
            if len(sample) > k:
                for j in range(k):
                    new_sample += sample[j]
                    new_sample += ' '
            else:
                for j in range(len(sample)):
                    new_sample += sample[j]
                    new_sample += ' '
    
            seqs.append(new_sample)
            labels.append(label)
        label += 1
```

```python
df = pd.DataFrame(list(zip(seqs, labels)), columns=['Sequence', 'Label'])
df.head(100)
```

```python
df.to_csv('Pdf_Json_1_DataFrame.csv', index=False)
```
